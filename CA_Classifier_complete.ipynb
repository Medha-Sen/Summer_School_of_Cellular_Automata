### Calculating other metrics such as Precision, Recall, F-1 score

```python
#we consider dataset1 to be positive class and dataset2 to be negative class
def metrics(dataset1,dataset2):
    true_positive=dataset1
    true_negative=dataset2
    #printing dataset1 and dataset2 values
    #print("Dataset1: ",dataset1)
    #print("Dataset2: ",dataset2)
    false_positive=0
    false_negative=0
   # print(misclassified)
    for i in misclassified:
        if i=="Dataset2.txt":
            false_negative=false_negative+1
        else:
            false_positive=false_positive+1
    precision= true_positive/(true_positive+false_positive)
    recall=true_positive/(true_positive+false_negative)
    f1_score=2*(precision*recall)/(precision+recall)
    return (precision,recall,f1_score)
```
